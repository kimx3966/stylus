```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import matplotlib

sns.set(style="whitegrid")
sns.set_context("paper")
sns.set_palette("colorblind")
matplotlib.rc('font', serif='cm10')
matplotlib.rc('mathtext', fontset='cm')
plt.rcParams['text.usetex'] = False
plt.rcParams["font.family"] = "Times New Roman"

categories = ['Visual Quality', 'Textual Alignment']
indexes = np.arange(len(categories))  # Numerical indexes for the x-axis positions
indexes = [0.15, 0.3]
bar_width = 0.1  # Width of the bars

lose = [100 * 16 / 84.0, 100 * (22) / float(100)]
tie = [100 * 44 / 84.0, 58]
win = [100 * 24 / 84.0, 100 * 20 / float(100)]
colors = sns.color_palette("colorblind", n_colors=10)

fig, ax = plt.subplots(figsize=(12, 14))

# Adjust the spacing here by modifying 'indexes' or the 'width'
lose_bars = ax.bar(indexes, lose, width=bar_width, label='Lose', color=colors[0], edgecolor='none')
tie_bars = ax.bar(indexes, tie, bottom=lose, width=bar_width, label='Tie', color=colors[1], edgecolor='none')
lose_tie_sum = [l + t for l, t in zip(lose, tie)]
win_bars = ax.bar(indexes, win, bottom=lose_tie_sum, width=bar_width, label='Win', color=colors[2], edgecolor='none')

# Adding text annotations inside bars
for bars in (lose_bars, tie_bars, win_bars):
    for bar in bars:
        height = bar.get_height()
        if height > 0:
            ax.annotate(f'{height:.0f}',
                        xy=(bar.get_x() + bar.get_width() / 2, bar.get_y() + height / 2 - 3),
                        xytext=(0, 3),  # 3 points vertical offset
                        textcoords="offset points",
                        ha='center', va='bottom', color='white', size=50)

ax.set_xticks(indexes)
ax.set_xticklabels([c.replace(" ", "\n") for c in categories], fontsize=50)  # Set the custom x-tick labels
ax.tick_params(axis='y', labelsize=32)
ax.set_ylabel('Preference Win Rate (%)', fontsize=50)
plt.ylim(top=100)
ax.legend(loc='upper center', frameon=False, bbox_to_anchor=(0.5, 1.12), ncol=3, fontsize=50)
plt.xlim(left=0.05, right=0.4)
plt.tight_layout()
plt.savefig('figures/gpt4_eval.pdf', format="pdf", dpi=500, bbox_inches='tight')
plt.show()
```
